```python
#Removing Outliers
from scipy.stats import zscore
z=np.abs(zscore(wine_data))
z
threshold=3
np.where(z>3)
```

```python
df=wine_data[(z<3).all(axis=1)]
df.shape
```

```python
print("Data Loss Percentage :",((wine_data.shape[0]-df.shape[0])/wine_data.shape[0])*100) 
```

 #data loss should be less than 10 %.

```python
x=df.drop('quality',axis=1)
y=df['quality']
```

```python
x
```

```python
y
```

```python
# Balancing the dataset using the SMOTE
import imblearn
from imblearn.over_sampling import SMOTE
x_smote,y_smote = SMOTE().fit_resample(x,y)
y_smote.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
max_Acc = 0
max_RS = 0

for i in range (1,200):
    x_train, x_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size = 0.20, random_state=i)
    lr = LogisticRegression()
    lr.fit(x_train,y_train)
    predlr = lr.predict(x_test)
    acc = accuracy_score(y_test, predlr)
    if acc>max_Acc:
        max_Acc = acc
        max_RS = i
        
print ('Best Accuracy is ', max_Acc, " on Random_State ", max_RS)
```

```python
#Now splitting the dataset into the Training set and test set
x_train,x_test,y_train,y_test=train_test_split(x_smote,y_smote,test_size=0.20,random_state=46)
```

```python
print('Size of X_train : ', x_train.shape)
print('Size of y_train : ', y_train.shape)
print('Size of X_test : ', x_test.shape)
print('Size of Y_test : ', y_test.shape)
```

```python
#feature scaling
st=StandardScaler()
x_train=st.fit_transform(x_train)
x_test =st.transform(x_test)
```

By doing Feature Scaling , we have scaled the data using standard scaler to overcome the issues of biasness.


```python
#Using PCA (Principal component analysis)for dimensionality reduction
pca=PCA(n_components=0.90)
x_train=pca.fit_transform(x_train)
x_test =pca.transform(x_test)
```

```python
sum(pca.explained_variance_ratio_)
```

Logistic Regression

```python
# Now applying Logistic regression
log= LogisticRegression()
log.fit(x_train,y_train)
y_pred1=log.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,y_pred1,output_dict=True))
print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,y_pred1)}\n")
print("Accuracy score",accuracy_score(y_test,y_pred1))
score = cross_val_score(log,x,y,cv=5)
print("Cross val score: ", score.mean())
```

Support Vector Classifier

```python
svm = SVC()
svm.fit(x_train,y_train)
y_pred2 = svm.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,y_pred2,output_dict=True))
print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,y_pred2)}\n")
print("Accuracy score",accuracy_score(y_test,y_pred2))
score1 = cross_val_score(svm,x,y)
print("Cross val score:",score1.mean())
```

DecisionTreeClassifier

```python
dt =DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred3= dt.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,y_pred3,output_dict=True))
print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,y_pred3)}\n")
print("Accuracy score",accuracy_score(y_test,y_pred3))
score2 = cross_val_score(dt,x,y)
print("Cross val score: ", score2.mean())
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred4 = rf.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,y_pred4,output_dict=True))
print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,y_pred4)}\n")
print("Accuracy score",accuracy_score(y_test,y_pred4))
score3 = cross_val_score(rf,x,y)
print("Cross val score: ", score3.mean())
```

GradientBoostingClassifier

```python
gbc= GradientBoostingClassifier()
gbc.fit(x_train,y_train)
y_pred5=gbc.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,y_pred5,output_dict=True))
print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,y_pred5)}\n")
print("Accuracy score:",accuracy_score(y_test,y_pred5))
score4 = cross_val_score(gbc,x,y)
print("Cross val score: ", score4.mean())
```

KNeighborsClassifier

```python
knn =KNeighborsClassifier()
knn.fit(x_train,y_train)
y_pred6 = knn.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,y_pred6,output_dict=True))
print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,y_pred6)}\n")
print("Accuracy score",accuracy_score(y_test,y_pred6))
score5 = cross_val_score(knn,x,y)
print("Cross val score: ", score5.mean())
```

Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
 
# Creating the hyperparameter grid
param_grid = {'solver': ['newton-cg', 'lbfgs', 'liblinear','saga','sag'],
              'penalty': ['l1','l2'],
              'C':[0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000],
             'dual':[True,False],
            'max_iter': [100,110,120,130,140]}
# Instantiating logistic regression classifier
model = LogisticRegression()
# Instantiating the GridSearchCV object
gsv = GridSearchCV( estimator = model,  param_grid = param_grid, cv = 5)
gsv.fit(x_train, y_train)
 
# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(gsv.best_params_))
print("Best score is {}".format(gsv.best_score_))
```
