```python
# Removing Skewness using power transformer
from sklearn.preprocessing import PowerTransformer
col = ['Mg','Ca','Ba','Fe','K']   # removing only skewed data

pt = PowerTransformer(method='yeo-johnson')
glass_df[col] = pt.fit_transform(glass_df[col].values)
```

```python
glass_df.skew()
```

```python
#Importing the libary Zscore to remove skewness
from scipy.stats import zscore
z=np.abs(zscore(glass_df))
z.shape
```

```python
threshold = 3
print(np.where(z>3))
```

```python
df= glass_df[(z<3).all(axis=1)]

print("Old DataFrame ",glass_df.shape)
print("New DataFrame ",df.shape)
print("Total Drop Rows are ",glass_df.shape[0]-df.shape[0])
```

```python
Data_Loss = print("Data Loss Percentage :",((glass_df.shape[0]-df.shape[0])/glass_df.shape[0])*100) 
```
